```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

uri = ("./data/dailypricehistory_kh.xlsx")
df = pd.read_excel(uri)
orig_df = df

df.index = df['Date']
df.index.to_period("D")
```

```python
df.head()
```

```python

#https://www.shanelynn.ie/summarising-aggregation-and-grouping-data-in-python-pandas/
#df.index.groupby(pd.Grouper(freq="M"))

def calculate(x):
    start_closing_price = x.loc[x.index.min(), "SPX"]
    end_closing_price = x.loc[x.index.max(), "SPX"]
    return end_closing_price-start_closing_price

result = df.groupby(df["Date"].dt.to_period("M")).apply(calculate)

result
```
